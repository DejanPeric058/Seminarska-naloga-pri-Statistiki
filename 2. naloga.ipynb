```python
import scipy.stats as staty
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
TempPulz = pd.read_csv('TempPulz.csv')
TempPulz
```

```python
temperature = TempPulz[['TEMPERATURA', 'SPOL']]
temperature
```

```python
def pretvori(stopinje_fahrenheita):
    return 5*(stopinje_fahrenheita-32)/9
def pretvori_std(stopinje_fahrenheita):
    return 5*(stopinje_fahrenheita)/9
```

```python
print('\n a) Ocenite povprečje in standardni odklon za telesno temperaturo posebej pri'
    'moških in posebej pri ženskah. \n')
n = 65
moški = temperature[temperature['SPOL'] == 1]['TEMPERATURA']
ženske = temperature[temperature['SPOL'] == 2]['TEMPERATURA']
povp_m = np.average(moški)
povp_ž = np.average(ženske)
povp_mC = pretvori(povp_m)
povp_žC = pretvori(povp_ž)
std_m = np.sqrt((n/(n-1))*staty.moment(moški, 2))
std_ž = np.sqrt((n/(n-1))*staty.moment(ženske, 2))
std_mC = pretvori_std(std_m)
std_žC = pretvori_std(std_ž)

print('Povprečje telesne temperature pri moških je {0}°F (oz. {1}°C), pri ženskah pa {2}°F (oz. {3}°C).'
      'Standardni odklon pri moških znaša {4}°F (oz. {5}°C), pri ženskah pa {6}°F (oz. {7}°C).'.format(povp_m,
    povp_mC, povp_ž, povp_žC, std_m, std_mC, std_ž, std_žC))
```

```python
print('\n b) Za povprečji iz prejšnje točke določite 95% intervala zaupanja. \n')
alpha = 0.05
normalna_inv = staty.norm.isf(alpha/2)
student_inv = staty.t.isf(alpha/2, n-1)
apr_m = std_m*normalna_inv/np.sqrt(n)
apr_ž = std_ž*normalna_inv/np.sqrt(n)
eks_m = std_m*student_inv/np.sqrt(n)
eks_ž = std_ž*student_inv/np.sqrt(n)
apr_m_zg = povp_m + apr_m
apr_m_sp = povp_m - apr_m
apr_ž_zg = povp_ž + apr_ž
apr_ž_sp = povp_ž - apr_ž
eks_m_zg = povp_m + eks_m
eks_m_sp = povp_m - eks_m
eks_ž_zg = povp_ž + eks_ž
eks_ž_sp = povp_ž - eks_ž
print('Aproksimativna intervala zaupanja: \n moški: ({0}°F, {1}°F) oz. ({2}°C, {3}°C)'
      '\n ženske: ({4}°F, {5}°F) oz. ({6}°C, {7}°C)'.format(apr_m_sp, apr_m_zg, pretvori(apr_m_sp), 
        pretvori(apr_m_zg), apr_ž_sp, apr_ž_zg, pretvori(apr_ž_sp), pretvori(apr_ž_zg)))
print('Eksaktna intervala zaupanja: \n moški: ({0}°F, {1}°F) oz. ({2}°C, {3}°C)'
      '\n ženske: ({4}°F, {5}°F) oz. ({6}°C, {7}°C)'.format(eks_m_sp, eks_m_zg, pretvori(eks_m_sp), 
        pretvori(eks_m_zg), eks_ž_sp, eks_ž_zg, pretvori(eks_ž_sp), pretvori(eks_ž_zg)))
```

```python
print('\n c) Preizkusite domnevo, da imajo moški in ženske v povprečju enako telesno temperaturo. \n')

se = np.sqrt((std_m*std_m + std_ž*std_ž)/n)
T = (povp_m - povp_ž)/se
abs_T = abs(T)

print('Vrednost testa je {0}, absolutna vrednost pa {1}'.format(T, abs_T))

alpha = 0.05
pogoj = staty.t.isf(alpha/2, n-1)
print(str(pogoj))
if abs_T < pogoj:
    print('Pri 0.05 hipotezo sprejmemo')
else:
    print('Pri 0.05 hipotezo ovržemo')
alpha = 0.01
pogoj = staty.t.isf(alpha/2, n-1)
print(str(pogoj))
if abs_T < pogoj:
    print('Pri 0.01 hipotezo sprejmemo')
else:
    print('Pri 0.01 hipotezo ovržemo')
print('ENOSTRANSKI TEST')
alpha = 0.05
pogoj = -staty.t.isf(alpha, n-1)
print(str(pogoj))
if T > pogoj:
    print('Pri 0.05 hipotezo sprejmemo')
else:
    print('Pri 0.05 hipotezo ovržemo')
alpha = 0.01
pogoj = -staty.t.isf(alpha, n-1)
print(str(pogoj))
if T > pogoj:
    print('Pri 0.01 hipotezo sprejmemo')
else:
    print('Pri 0.01 hipotezo ovržemo')
    
```
